```python
#importing packages
import torch
import numpy as np
import torch.nn as nn
import torchvision
from torch.utils.data import Dataset
import numpy as np
import pandas as pd
```

```python
#creating custom dataset
data=pd.read_csv(r"/content/nse.csv")
data=data.dropna()

print(len(data))
print(type(data))
```

```python
#custom datasetclass
class Mydataset(Dataset):
  def __init__(self,data):
    super(Mydataset,self).__init__()
    self.data=data.iloc[:,3:7]


  def __len__(self):
    return len(self.data)-365


  def __getitem__(self,idx):

    p=np.array(self.data.iloc[idx:idx+365,])
    q=np.array(self.data.iloc[idx+365] )         
    x_value=torch.tensor(p,dtype=torch.float32)
    x_value=x_value.reshape(-1,)
    y_value=torch.tensor(q,dtype=torch.float32)
    return x_value,y_value






```

```python
#configuring device
device='cuda' if torch.cuda.is_available() else 'cpu'
```

```python
#defining hyperparameters
batch_size=5
num_epochs=100
learning_rate=0.001
input_size=1460
output_size=4
```

```python
#loading data
train_dataset=Mydataset(data)
train_loader=torch.utils.data.DataLoader(dataset=train_dataset,batch_size=batch_size,shuffle=True)
```

```python
#define network
class network(nn.Module):
  def __init__(self,input_size,output_size):
    super(network,self).__init__()
    self.layer1=nn.Linear(input_size,365)
    self.layer2=nn.Linear(365,120)
    self.layer3=nn.Linear(120,30)
    self.layer4=nn.Linear(30,output_size)
    self.relu=nn.ReLU()
  def forward(self,x):
    x=self.layer1(x)
    x=self.layer2(x)
    x=self.relu(x)
    x=self.layer3(x)
    x=self.relu(x)
    out=self.layer4(x)
    return out
model=network(input_size,output_size).to(device)   
```

```python
#define loss func and optim
criterion=nn.L1Loss()
optim=torch.optim.Adam(params=model.parameters(),lr=learning_rate)
```

```python
#training
for epoch in range(num_epochs):
  for i,(image,labels) in enumerate(train_loader):
    image=image.reshape(-1,1460).to(device)
    labels=labels.to(device)
    optim.zero_grad()
    pred=model(image)
    loss=criterion(pred,labels)
    loss.backward()
    optim.step()
    if i%100==0:
      print("epoch[{}/{}] loss:[{}]".format(epoch,num_epochs,loss.item()))
```
